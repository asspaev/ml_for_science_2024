## Plotly

Plotly -  библиотека для интерактивной визуализации данных

[Документация](https://plotly.com/python/)



```python
import plotly.express as px
import pandas as pd
```

#### Sunburst plot - визуализация иерархических данных

```python
data = dict(
    categories = ['Total', 'A', 'B', 'C',
                  'A1', 'A2', 'A3',
                  'B1', 'B2',
                  'C1', 'C2', 'C3'],
    parent = ['', 'Total', 'Total', 'Total', 'A', 'A', 'A', 'B', 'B', 'C', 'C', 'C'],
    value = [100, 35, 25, 40, 10, 15, 10, 12, 13, 10, 20, 10]
)

fig = px.sunburst(data, names='categories', parents='parent', values='value', branchvalues='total')
fig.show()
```

#### Heatmap

```python
data = [[5, 23, 15, 47], [25, 5, 25, 12],
        [31, 35, 25, 41], [12, 18, 5, 7]]

fig = px.imshow(data, color_continuous_scale='RdYlBu', x=['A', 'B', 'C', 'D'], y=['V1', 'V2', 'V3', 'V4'])
fig.show()
```

#### Funnel chart - воронка

```python
df = pd.DataFrame(dict(
    stage = ['Sent', 'Received', 'Clicks', 'Sales', 'Sent', 'Received', 'Clicks', 'Sales'],
    percentage = [70, 32, 20, 4, 30, 38, 10, 6],
    campaign = ['C1', 'C1', 'C1', 'C1', 'C2', 'C2', 'C2', 'C2']))
df.head()
```

```python
df = pd.DataFrame(dict(
    stage = ['Sent', 'Received', 'Clicks', 'Sales', 'Sent', 'Received', 'Clicks', 'Sales'],
    percentage = [70, 32, 20, 4, 30, 38, 10, 6],
    campaign = ['C1', 'C1', 'C1', 'C1', 'C2', 'C2', 'C2', 'C2']))

fig = px.funnel(df, x='percentage', y='stage', color='campaign')
fig.show()
```

#### Bubble plot - scatter plot с точками разного размера

```python
fig = px.scatter(x=[5, 7, 3, 6, 1, 9, 2], y=[2, 6, 2, 1, 4, 1, 6], size=[30, 25, 250, 12, 140, 80, 36])
fig.show()
```

#### Heatmap на карте

Набор данных представляет собой данные переписи населения Галисии - автономного сообщества Испании. Координаты - центр каждого участка переписи.

```python
df = pd.read_csv('https://raw.githubusercontent.com/R-CoderDotCom/data/main/sample_datasets/population_galicia.csv')
df.head()
```

newplot.png

```python
fig = px.density_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    z='tot_pob',
    radius=7,
    center=dict(lat=42.83, lon=-8.35),
    zoom=6,
    mapbox_style='open-street-map',
    color_continuous_scale='rainbow',
    opacity=0.5,
)
fig.show()
```

```python
fig.write_html("map.html")
```
